```python
# Import 
import pandas as pd
import numpy as np

# Show all columns
pd.set_option('display.max_columns', None)

####### pip install openpyxl
# Read xlsx
df_FP2 = pd.read_excel('FP2_HF_DataValues.xlsx')
# Read Data.csv
data = pd.read_csv('C:/Users/mathi/OneDrive/Skrivebord/2_Sem_Kand/Projekt_Kandidat2/DataPrep.csv')

## Print all DataTypeLabel in df_FP2
# df_FP2['DataTypeLabel'].unique() # 'Blodtryk, Systolisk', 'Blodtryk, Diastolisk', 'Puls', 'Vægt', 'Skridt'

df_FP2
```

```python
import datetime


# Take dataframes data and make new column 'from_datetime' to convert unix emffit that is in the format 1602086420 to datetime
data['from_datetime'] = pd.to_datetime(data['from'], unit='s')

# convert 'TimeMeasrd' column to datetime format
df_FP2['TimeMeasrd'] = pd.to_datetime(df_FP2['TimeMeasrd'])

# safe unique values from 'DataTypeLabel' in a list
DataTypeLabel = df_FP2['DataTypeLabel'].unique()

# Make new column names for each DataTypeLabel
for i in DataTypeLabel:
    data[i] = np.nan


# Now use from data['from_datetime'] to find the right row in df_FP2 and fill in the right value in the right column
for i in range(len(data)):
    print(i)
    for j in DataTypeLabel:
        try:
            # Indsæt i data dataframe 'Values' fra df_FP2 dataframe hvis 'from_datetime' er inden for den samme dag som 'TimeMeasrd'
            # og hvis SubjectId i df_FP2 matcher patient_ID i data dataframe
            data.loc[i, j] = df_FP2.loc[(df_FP2['DataTypeLabel'] == j) & (pd.to_datetime(df_FP2['TimeMeasrd']).dt.date == pd.to_datetime(data['from_datetime'][i]).date()) & (df_FP2['SubjectId'] == data['patient_ID'][i]), 'Value'].iloc[0]
        except IndexError:
            print(f"Ingen værdi fundet for betingelsen for række {i} og kolonne {j}")

data
```

```python
# Drop kolonne 'user_id'
data.drop('user_id', axis=1, inplace=True)

# # Tjek alle kolonner for NaN værdier
# data.isna().sum()

# Gem data som csv 
data.to_csv('DataPrep.csv', index=False)
```
