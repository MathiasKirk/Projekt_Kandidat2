```python

# Make a list of unique values in the column 'subject_ID'
unique_subject_ID = df_patientInfo['subject_ID'].unique()
# print(unique_subject_ID) # [431 435 440 448 451 453 454 457 459 462 465 468 469 477 478 483 485 492
#  494 495 499 504 510 513 515 516 518 520 524 526 539 545 548 552 554 586
#  583 593 580 579 568 560 571 563 569 584 564 592 597 601 606 610 612 620
#  623]

# Print unique values in the column 'patient_ID'
unique_patient_ID = data['patient_ID'].unique() 
# print(unique_patient_ID)  # [152 258 261 264 271 283 285 286 294 299 300 301 318 323 327 329 332 342
#  350 352 353 360 361 376 383 386 402 403 404 405 406 407 408 409 410 411
#  412 413 414 415 416 417 418 419 420 421 431 435 440 448 451 454 459 462
#  465 468 469 477 485 492 494 495 504 510 513 515 516 520 524 526 545 548
#  552 554 560 563 564 569 571 583 586 579 580 584 592 593 597 601 606 610
#  612 620 623]

# # Which values in unique_patient_ID are not in unique_subject_ID
# print(set(unique_patient_ID) - set(unique_subject_ID))

# Only keep the values in data['patient_ID'] that are in unique_subject_ID
data = data[data['patient_ID'].isin(unique_subject_ID)]


# # For loop through the list and where unique_subject_ID is == patient_ID in dataframe data then make new column age_baseline	gender	NYHA_baseline	ef_titration	ef_rehabilitation	ef_finished	weight_baseline	height_baseline	bp_sys_baseline	bp_dia_baseline	heart_rate_baseline and add the values from the dataframe df_patientInfo

for i in unique_subject_ID:
    data.loc[data['patient_ID'] == i, 'AGEBL'] = df_patientInfo.loc[df_patientInfo['subject_ID'] == i, 'age_baseline'].values[0]
    data.loc[data['patient_ID'] == i, 'GENDER'] = df_patientInfo.loc[df_patientInfo['subject_ID'] == i, 'gender'].values[0]
    data.loc[data['patient_ID'] == i, 'NYHABL'] = df_patientInfo.loc[df_patientInfo['subject_ID'] == i, 'NYHA_baseline'].values[0]
    data.loc[data['patient_ID'] == i, 'EFTITBL'] = df_patientInfo.loc[df_patientInfo['subject_ID'] == i, 'ef_titration'].values[0]
    data.loc[data['patient_ID'] == i, 'EJREHABBL'] = df_patientInfo.loc[df_patientInfo['subject_ID'] == i, 'ef_rehabilitation'].values[0]
    data.loc[data['patient_ID'] == i, 'EJFINISHBL'] = df_patientInfo.loc[df_patientInfo['subject_ID'] == i, 'ef_finished'].values[0]
    data.loc[data['patient_ID'] == i, 'WEIGHTBL'] = df_patientInfo.loc[df_patientInfo['subject_ID'] == i, 'weight_baseline'].values[0]
    data.loc[data['patient_ID'] == i, 'HEIGHTBL'] = df_patientInfo.loc[df_patientInfo['subject_ID'] == i, 'height_baseline'].values[0]
    data.loc[data['patient_ID'] == i, 'BPSYSBL'] = df_patientInfo.loc[df_patientInfo['subject_ID'] == i, 'bp_sys_baseline'].values[0]
    data.loc[data['patient_ID'] == i, 'BPDIABL'] = df_patientInfo.loc[df_patientInfo['subject_ID'] == i, 'bp_dia_baseline'].values[0]
    data.loc[data['patient_ID'] == i, 'HRB'] = df_patientInfo.loc[df_patientInfo['subject_ID'] == i, 'heart_rate_baseline'].values[0]

# data

# # # # Print data for patient_ID == 524
# # # data.loc[data['patient_ID'] == 524]

# Save as new csv file
data.to_csv('Data2.csv', index=False)

```
