### Sbt152

```python
# Import 
import pandas as pd
import numpy as np
import os

# Show all columns
pd.set_option('display.max_columns', None)

# Define the directory where the log files are located
log_dir = "C:/Users/mathi/OneDrive/Skrivebord/2_Sem_Kand/Projekt_Kandidat2/Emfit/Sbt152/"

# Define the desired column names
desired_columns = ['id', 'device', 'user_id', 'from', 'to', 'avg_hr', 'avg_rr', 'avg_act',
        'calc_data', 'sleep_score', 'duration', 'duration_in_bed',
        'duration_awake', 'duration_in_sleep', 'duration_in_rem',
        'duration_in_light', 'duration_in_deep', 'duration_sleep_onset',
        'bedexit_duration', 'sleep_data', 'bedexit_count', 'bedexit_data',
        'tossnturn_count', 'tossnturn_data', 'from_gmt_offset', 'min_hr',
        'max_hr', 'min_rr', 'max_rr', 'fm_count', 'fm_data', 'awakenings',
        'hrv_data', 'hrv_score', 'hrv_lf', 'hrv_hf', 'hrv_rmssd_data',
        'hrv_rmssd_evening', 'hrv_rmssd_morning', 'temp_v1_uid']

# Create an empty dataframe with the desired column names
df = pd.DataFrame(columns=desired_columns)

# Loop over all the log files in the directory
for filename in os.listdir(log_dir):
    if filename.endswith(".log"):
        file_path = os.path.join(log_dir, filename)
        # print(f"Processing file {file_path}")
        
        # Read the log file into a temporary dataframe
        data = pd.read_json(file_path, lines=True)
        temp_df = pd.DataFrame(data)
        
        # Check if the temporary dataframe has all the desired columns
        if set(desired_columns).issubset(set(temp_df.columns)):
            # If it does, append it to the main dataframe
            df = pd.concat([df, temp_df[desired_columns]], ignore_index=True)

        else:
            # If it doesn't, create a new dataframe with only the columns that exist in the temporary dataframe
            # and append it to the main dataframe
            # print(f"File {file_path} has missing columns")
            missing_columns = set(desired_columns) - set(temp_df.columns)
            new_columns = [col for col in desired_columns if col not in missing_columns]
            df = pd.concat([df, temp_df[new_columns]], ignore_index=True)

# Display the resulting dataframe
df1 = df
df1.head(5)
```
